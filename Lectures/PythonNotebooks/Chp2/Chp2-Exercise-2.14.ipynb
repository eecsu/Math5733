```python
# Part (c)

n_trials = [5, 10, 20, 40, 80]
E_h = np.zeros(5)
h = np.zeros(5)
alpha_h = np.zeros(5)

count = 0
print '   n    &    h     &      E_h     &   Rate of Conv. ' + r'\\ \hline'
for n in n_trials:
    A = np.zeros((n+1,n+1))
    np.fill_diagonal(A,2)
    A += np.diag(-np.ones(n),k=1)
    A += np.diag(-np.ones(n),k=-1)
    A[-1,-2] += -1
    
    x = np.linspace(0,1,n+2)
    h[count] = x[1]-x[0]
    b = h[count]**2 * -np.exp(x-1) # <------- This line needs to change so the numerical solution is correct
    b[-1] += 2*h[count]
    
    v = np.zeros(n+2)
    v[1:] = np.linalg.solve(A, b[1:]) #Numerical soln. using Gaussian elimination

    u = np.exp(-1)*(np.exp(x)-1) #Exact soln. <------- This line needs to change so that the exact solution is correct
    
    E_h[count] = np.max(np.abs(u-v))
    
    if count == 0:
        print ' %3d    &   %.3f  & %.8f   &         ' % (n, h[count], E_h[count]) + r'   \\'
    else:
        alpha_h[count] = np.log(E_h[count]/E_h[count-1])/np.log(h[count]/h[count-1])
        print ' %3d    &   %.3f  & %.8f   &   %.4f ' % (n, h[count], E_h[count], alpha_h[count]) + r'  \\'
    
    
    plt.figure(count)
    plt.plot(x,v,label='Num. Soln. $v$')
    plt.plot(x,u,label='Exact Soln. $u$')
    plt.legend(loc='upper left', shadow=True)

    count +=1
    
    
```
